```python
#pip install BeautifulSoup4
#pip install lxml

import requests
from bs4 import BeautifulSoup
import csv

filename = '덕양구 군 우대업소 정보.csv'
f = open(filename, "w", encoding="utf-8-sig", newline="")
writer = csv.writer(f)

headers = {
    'Uesr-Agent':
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'
}
for i in range(1,17):
  url = 'http://www.goyang.go.kr/dygu/sldrPrfrBsns/BD_selectSldrPrfrBsnsList.do?q_searchKeyTy=&q_searchVal=&q_rowPerPage=4&q_currPage={}&q_sortName=&q_sortOrder=&'.format(
        i)
  res = requests.get(url)
  res.raise_for_status()
  soup = BeautifulSoup(res.text, 'lxml')

  columns = soup.find_all("td")
  data = [column.get_text() for column in columns]
  data2 = [data[i:i+5] for i in range(0, len(data), 5)]
  for j in range(len(data2)):
    data3 = data2[j]
    writer.writerow(data3)
```
